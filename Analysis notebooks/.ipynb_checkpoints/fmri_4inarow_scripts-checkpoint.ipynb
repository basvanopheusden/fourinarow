```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.colors as colors
import pandas as pd
```

```python
cm = colors.LinearSegmentedColormap.from_list('gray_gold_map', [colors.to_rgb('darkgray'), 
                                                                colors.to_rgb('gold')], N=100)
def show_board(bp,wp,response,color,save=False):    
    fig = plt.figure(figsize=[9,4])
    ax = fig.add_subplot(111,aspect='equal')
    ax.vlines(np.arange(-0.5,9.5,1),-0.5,3.5)
    ax.hlines(np.arange(-0.5,4.5,1),-0.5,8.5)
    
    black_pieces = np.nonzero(np.array(list(bp)).astype(int))[0]
    white_pieces = np.nonzero(np.array(list(wp)).astype(int))[0]
    
    for p in black_pieces:
        circ = patches.Circle((p%9,p//9),0.33,color="black",fill=True)
        circ = ax.add_patch(circ)
    for p in white_pieces:
        circ = patches.Circle((p%9,p//9),0.33,color="white",fill=True)
        circ = ax.add_patch(circ)
    for p in [response]:
        circ = patches.Circle((p%9,p//9),0.33,color=color,fill=False)
        circ = ax.add_patch(circ)
    plt.imshow(np.zeros(shape=[4,9]), cmap=cm, 
               interpolation='nearest',origin='bottom',vmin=0,vmax=0.2)
    ax.axis('off')
    fig.tight_layout()
    if save:
        fig.savefig('C:/Users/svo/Documents/fmri/Boards/board_' + bp + '_' + wp + '.png')
    plt.show()
```

```python
def encode_pieces(pieces):
    return np.sum([2**i for i,p in enumerate(pieces) if p=='1'])

def encode_color(c):
    return "Black" if c=="black" else "White"

def encode_response(r):
    return 2**r if r in range(36) else 0
    
def save_valid_boards(df):
    df_new = pd.DataFrame(zip([encode_pieces(bp) for bp in df['bp'].values],
                     [encode_pieces(wp) for wp in df['wp'].values],
                     [encode_color(c) for c in df['color'].values],
                     [encode_response(r) for r in df['response']],
                     np.zeros(len(df)),
                     df['name']))

    df_new = df_new[df_new[3]>0]
    for name in np.unique(df_new[5]):
        with open(name + '.csv','w') as f:
            df_new[df_new[5]==name].to_csv(f,index=False,header=False,line_terminator='\n')
            
def save_invalid_boards(df,participant_dict):
    df_new = pd.DataFrame(zip([encode_pieces(bp) for bp in df['bp'].values],
                     [encode_pieces(wp) for wp in df['wp'].values],
                     [encode_color(c) for c in df['color'].values],
                     [encode_response(r) for r in df['response']],
                     np.zeros(len(df)),
                     np.zeros(len(df)).astype(int),
                     [participant_dict[name] for name in df['name']]))

    df_new = df_new[df_new[3]==0]
    with open(direc + 'invalid_boards.csv','w') as f:
        df_new.to_csv(f,index=False,header=False,sep='\t',line_terminator='\n')
```

```python
def expand_params(theta):
    thresh = theta[0];
    delta = theta[2];
    w_center = theta[5];
    w = theta[6:10]
    lapse = theta[3];
    c_act = theta[4];
    gamma = theta[1];
    return np.hstack([[10000,thresh,gamma,lapse,1,1,w_center],
                     np.tile(w,4),
                     [0],
                     c_act*np.tile(w,4),
                     [0],
                     [delta]*17])
    
```

```python
params_short = np.vstack([np.loadtxt(direc + '/splits/' + str(i+1) + '/params' + str(0) + '.csv',delimiter=',') 
           for i in range(len(behavior_names)) for g in range(1,6)])

params = np.vstack([expand_params(np.loadtxt(direc + '/splits/' + str(i+1) + '/params' + str(0) + '.csv',delimiter=',')) 
           for i in range(len(behavior_names)) for g in range(1,6)])
    
#np.savetxt(direc + 'params_fmri_final_long.txt',params,fmt = '%6f')

v_3inarow = 0.4
```

```python
for name,vals in zip(['entropy','value_conditioned','value_myopic','depth','iterations'],regressors):
    
    boards[name] = np.full(len(boards),fill_value=np.nan)

    for participant_name,num in participant_dict.items():
        if 'value' in name:
            r=np.mean(np.tanh(v_3inarow*vals[num]/params_short[5*num,-2]),axis=1)
        elif name=='depth':
            r=np.mean(np.maximum(vals[num]-1,0),axis=1)
        elif name=='iterations':
            r=np.mean(vals[num],axis=1)
        else:
            r = vals[num]
        boards.loc[(boards['response']!=99) & (boards['name']==participant_name),name] = r#vals[num]
    
#for name,vals in zip(['value','myopic_value'],
#                    [fmri_board_values_invalid,fmri_board_values_invalid_notree]):
#    for participant_name,num in participant_dict.items():
#        boards.loc[(boards['response']==99) & (boards['name']==participant_name),name] = vals[invalid_boards_cpp[6]==num]

for name in ['value_myopic','value_conditioned']:
    boards[name] *= (2*(boards['color']=='black')-1)
```

```python
import scipy.stats as sts

def my_corr(x,y):
    ind = np.logical_and(np.isnan(x),np.isnan(y))
    return sts.pearsonr(x[~ind],y[~ind])[0]

for key in ['entropy','value_conditioned','value_myopic','depth','iterations']:
    plt.plot(boards[key],old_boards[key],'.',alpha=0.25)
    print(key, np.mean([my_corr(boards[boards['name']==name][key],old_boards[old_boards['name']==name][key]) 
                for name in np.unique(boards['name'])]))
    plt.show()
```
